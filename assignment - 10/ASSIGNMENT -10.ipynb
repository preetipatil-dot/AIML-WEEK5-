## I. Univariate Analysis for Continuous Variables and Categorical Variables

## II. Bivariate Analysis for Continuous Variable vs Continuous Variable, Categorical Variable vs Categorical Variable**

```python
#Finding the correlation of variables such as Age, Debt, YearsEmployed, CreditScore, In
df[['Age','Debt','YearsEmployed','CreditScore','Income']].corr()
```

```python
# Plotting a scatter plot of YearsEmployed,Income
from matplotlib import pyplot as plt
sns.scatterplot(df.YearsEmployed,df.Income)
plt.ylim(0,2000)
```

## Bivariate Analysis of Categorical Variable vs ContinousVariable

```python
# Getting an average credit score of the people on the basis of Approval
df.groupby(by='Approved').agg('mean')[['Age','Debt','YearsEmployed','CreditScore']]
```

```python
# Plotting a kde plot with hue=Approved
plt.figure(figsize=(10,8))
sns.kdeplot(data=df,x='Age',hue='Approved',fill=True)
```

## Bivariate Analysis of Categorical Variable vs Categorical Variable

```python
#plotting a count plot of a categorical variable with the hue= 'Gender'
#sns.countplot(df,x='Approved',hue='Gender')
# Getting the percentages of gender getting approval
all=pd.crosstab(df.Gender,df.Approved,margins=True)['All']
pd.crosstab(df.Gender,df.Approved).divide(all,axis=0).dropna()
```

## III.Multivariate Analysis for Numerical-Numerical-Categorical Variables


```python
#Plotting a pairplot of all variables with the hue = 'Approved'
sns.pairplot(data=df[['Age','Debt','YearsEmployed','CreditScore','Income', 'Approved']],hue='Approved')
```
