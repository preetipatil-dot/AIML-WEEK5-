```python
#load dependencies
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
titanic=pd.read_csv("titanic_data.csv")
print(titanic.shape)
titanic
```

```python
titanic.describe()
```

## I.Total number of passengers from each class who survived

```python
titanic.groupby(['class'])['survived'].sum().reset_index()
```

Note:The reset_index() helps bring the grouping columns from index,back as a column in
dataframe.

## II.Average fare for passengers from each town

```python
titanic.groupby(['embark_town'])['fare'].mean().reset_index()
```

```python
print(titanic['who'].nunique())
```

## III. What was the average fare for male vs female passengers from each town? (adding more groups/levels)


```python
titanic.groupby(['embark_town','sex'])['fare'].mean().reset_index()
```

## IV. What was the average fare and age for male vs female passengers from each town? (Adding more variables/features)


```python
titanic.groupby(['embark_town','sex'])['fare','age'].mean().reset_index()
```

## V.Get the unique set of ages for all each age category (who column) from each town


```python
titanic.groupby(['embark_town','who'])['age'].apply(set).reset_index()
```

## VI.Get the range (min - max) of ages for each age category (who column) from each town

```python

titanic.groupby(['embark_town','who'])['age'].apply(lambda x:x.max()).reset_index()
```

## VII. Get the mean fare by age ratio for each age category (who column) from each town.

```python
titanic.groupby(['embark_town','who']).apply(lambda x: (x['fare']/x['age']).mean())  
```

## VIII.Get the mean of fare, AND median of age for each age category (who column) from each town


```python
#Define aggregations as a dictionary
x = {'fare':'mean','age':'median'}
titanic.groupby (['embark_town', 'who']).agg(x).reset_index()
```

## IX.Get the sum & mean of fare, AND median, min and max of age for each age category (who column) from each town

```python
x={'fare':['sum','mean'],'age':['median','min','max']}
titanic.groupby(['embark_town','who']).agg(x).reset_index()
```

## x) Create a new column that returns the average fare for the age group (who column) the passenger belongs to.

```python
x={'fare':sum,'age':max}
titanic.groupby(['embark_town','who']).agg(x).reset_index()
```
