```python
#Visualizing the species count
plt.title('Species count')
sns.countplot(df['Species'])
```

```python
#Comparison between sepal width and length n the basis of species
plt.figure(figsize=(16,9))
plt.title('Comparison between sepal width and length n the basis of species')
sns.scatterplot(df['SepalLengthCm'],df['SepalWidthCm'],hue=df['Species'])
```

```python
#Comparison between petal width and length
plt.title("Comparison between petal width and length")
sns.scatterplot(df['PetalLengthCm'],df['PetalWidthCm'])
```

```python
#Comparison between sepal width and length n the basis of species
plt.figure(figsize=(16,9))
plt.title('Comparison between sepal width and length n the basis of species')
sns.scatterplot(df['PetalLengthCm'],df['PetalWidthCm'],hue=df['Species'],s=50)
```
